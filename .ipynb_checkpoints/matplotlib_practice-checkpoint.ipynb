```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot (MATLAB STYLE)
plt.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="red", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt() [OBJECT-ORIENTED]
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots() #(figsize=(w, h))#Change the size of the figure in pixels. Add this in the subplots() function.
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')#Add a label that will be added to the legend.
#ax.set_xlim(min, max) #Sets the min and max range of the x-axis.
ax.set_ylim(0, 45) #Sets the min and max range of the y-axis.
ax.set_xlabel('Monthsl') #Add a label to the x-axis.
ax.set_ylabel('Fare ($)') #Add a label to the y-axis.
ax.set_title("PyBer Fare by Month") #Add a title.
ax.legend() #Add a legend.
ax.grid() #Add a grid to the chart.
# ** plt.savefig("add a path and figure extension")
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="cyan", label="Chicago")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.xlim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c='red', label="Chicago")
plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0, 45)
```

```python

#y_axis_larger = []
#for data in y_axis:
#    y_axis_larger.append(data*3)
plt.scatter(y_axis, x_axis, c='red', label="Chicago", s = [i * 3 for i in y_axis])
plt.ylabel("Fare ($)")
plt.xlabel("Month")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0, 45)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')

```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
